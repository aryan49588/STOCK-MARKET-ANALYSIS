```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stock=pd.read_csv('nifty_500.csv')
stock
```

```python
stock.sort_values('High',ascending=False)
```

```python
sns.relplot(x="High",y="Company Name",data=stock.sort_values('High',ascending=False).head(5))
plt.show()
```

```python
#Top five gainers are MRF LTD.,Page Industries Ltd,Honeywell Automation India Ltd,3M INDIA LTd and Shree Cement LTD. 
```

```python
print("Top ten companies based on Last Traded Price.....")
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="Last Traded Price",y="Company Name",data=stock.sort_values('High',ascending=False).head(10),hue='Company Name')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,8)})#Lineplot for Same data
sns.lineplot(x="Last Traded Price",y="Company Name",data=stock.sort_values('High',ascending=False).head(10),color='Orange')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.relplot(x="Last Traded Price",y="Company Name",data=stock.sort_values('High',ascending=False).head(10))
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.histplot(x="Last Traded Price",y="Company Name",data=stock.sort_values('High',ascending=False).head(10))
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.jointplot(x="Last Traded Price",y="Company Name",data=stock.sort_values('High',ascending=False).head(10))
plt.show()
```
