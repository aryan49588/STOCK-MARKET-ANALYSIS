```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mydata=pd.read_csv('raw_data.csv')
mydata
```

```python
mydata.count()
```

```python
mydata['date'] = pd.to_datetime(mydata['date'], format='%Y-%m-%d')
```

```python
mydata
```

```python
filtered_mydata=mydata.loc[(mydata['date'] >= '2020-09-01') 
                       & (mydata['date'] < '2020-09-02')]
filtered_mydata
```

```python
sorted_data=filtered_mydata.sort_values('total_cases',ascending=False)
sorted_data
```

```python
print("Top ten countries with highest number of total cases as on 2020-09-01")
sorted_data.head(10)
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="location",y="total_cases",data=sorted_data.head(10),hue="location")
plt.show()
```

```python
#For example we can analyse covid 19 impact for particular country ,say, Mexico
```

```python
mexico=mydata[mydata['location']=='Mexico']
mexico
```

```python
mex_data=mexico.loc[(mexico['date']>='2020-10-01')
          & (mexico['date'] <='2020-10-30')]
mex_data
```

```python
print("Rising cases in MEXICO...............")
sns.set(rc={'figure.figsize':(15,8)})
sns.lineplot(x='total_cases',y='date',data=mex_data,color='red')
plt.show()
```

```python
#For example we can analyse covid 19 impact for particular country ,say, Sweden
```

```python
sweden=mydata[mydata['location']=='Sweden']
sweden
```

```python
swe_data=sweden.loc[(sweden['date']>='2020-09-01')
          & (sweden['date'] <='2020-09-30')]
swe_data
```

```python
print("Rise in total covid cases in Sweden from date 2020-09-01 to 2020-10-01")

sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="total_cases",y="date",data=swe_data,hue='location')
plt.show()
```

```python
print("Rise in total covid cases in Sweden from date 2020-09-01 to 2020-10-01")

sns.lineplot(x="total_cases",y='date',data=swe_data,color='Green')
plt.show()
```

```python
sns.relplot(x="total_cases",y='date',data=swe_data)
plt.show()
```
